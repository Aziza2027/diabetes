```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder

# Create sample data with categorical columns
df = pd.DataFrame({'A': ['a', 'b', 'c', 'd', 'e'],
                   'B': ['x', 'y', 'z', 'x', 'y'],
                   'C': ['high', 'low', 'medium', np.none, 'low']})

# Instantiate the OrdinalEncoder and fit-transform the data
encoder = OrdinalEncoder()
df_encoded = pd.DataFrame(encoder.fit_transform(cat_data), columns=cat_data.columns)

# Print the encoded dataframe
print('Encoded dataframe:\n', df_encoded)

# Inverse transform the encoded data to get the original data
df_decoded = pd.DataFrame(encoder.inverse_transform(df_encoded), columns=df.columns)

# Print the decoded dataframe
print('\nDecoded dataframe:\n', df_decoded)
```

```python

# list of columns for which to calculate odds ratio
genotype_cols = [col for col in data.columns if col != 'case']
```

```python

# loop over genotype columns
for col in genotype_cols:
    # split into case and control groups
    case_group = data[data['case'] == 1][col]
    control_group = data[data['case'] == 0][col]

    # extract wild type and mutant type from column name
    w, m = col[-3:].split('>')
    wild, heterozygous, mutant = w+w, w+m, m+m

    # calculate odds ratio and confidence interval for codominant model
    odds_ratio_codom, lower_ci_codom, upper_ci_codom = odds_ratio(
        np.array([[sum(case_group == wild), sum(case_group == heterozygous), sum(case_group == mutant)],
                  [sum(control_group == wild), sum(control_group == heterozygous), sum(control_group == mutant)]])
    )

    # calculate odds ratio and confidence interval for dominant model
    odds_ratio_dom, lower_ci_dom, upper_ci_dom = odds_ratio(
        np.array([[sum(case_group == heterozygous) + sum(case_group == mutant),
                   sum(case_group == wild)],
                  [sum(control_group == heterozygous) + sum(control_group == mutant),
                   sum(control_group == wild)]])
    )

    # calculate odds ratio and confidence interval for recessive model
    odds_ratio_rec, lower_ci_rec, upper_ci_rec = odds_ratio(
        np.array([[sum(case_group == mutant),
                   sum(case_group == wild) + sum(case_group == heterozygous)],
                  [sum(control_group == mutant),
                   sum(control_group == wild) + sum(control_group == heterozygous)]])
    )

    # print results
    print(f"Odds ratio for {col} (codominant model): {odds_ratio_codom:.2f}, CI [{lower_ci_codom:.2f}, {upper_ci_codom:.2f}]")
    print(f"Odds ratio for {col} (dominant model): {odds_ratio_dom:.2f}, CI [{lower_ci_dom:.2f}, {upper_ci_dom:.2f}]")
    print(f"Odds ratio for {col} (recessive model): {odds_ratio_rec:.2f}, CI [{lower_ci_rec:.2f}, {upper_ci_rec:.2f}]")
```

```python

import pandas as pd
from sklearn.impute import KNNImputer
from sklearn.preprocessing import OneHotEncoder

# Create sample data with missing values
df = pd.DataFrame({'A': ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k'],
                   'B': ['x', 'y', 'z', None, 'x', 'y', None, 'z', 'y', None, 'x']})

# One-hot encode the categorical variables
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
df_encoded = pd.DataFrame(encoder.fit_transform(df[['A', 'B']]))

# Instantiate the KNN imputer
imputer = KNNImputer(n_neighbors=2)

# Fit and transform the encoded data
df_encoded_filled = pd.DataFrame(imputer.fit_transform(df_encoded))

# Inverse transform the encoded data to get the original categorical variables
df_filled = pd.DataFrame(encoder.inverse_transform(df_encoded_filled))

# Rename the columns of the filled data
df_filled.columns = ['A', 'B']

# Print the original data and the filled data
print('Original data:\n', df)
print('\nFilled data:\n', df_filled)
```
